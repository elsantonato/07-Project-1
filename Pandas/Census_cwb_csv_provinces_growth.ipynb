```python
# Import Dependencies
import pandas as pd
import numpy as np

# Make a reference to the books.csv file path
path_2011="../Input/cwb_2011.csv"

# Import the books.csv file as a DataFrame
cwb2011_df=pd.read_csv(path_2011, encoding="ISO-8859-1")

cwb2011_df.head()

```

```python
cwb2011_df["CSD Code / Code de la SDR 2011"]=cwb2011_df["CSD Code / Code de la SDR 2011"].astype(str)
cwb2011_df["Province Code"]=cwb2011_df["CSD Code / Code de la SDR 2011"].str[0:2]
cwb2011_df.head()
```

```python
# Import the books.csv file as a DataFrame
path_PC="../Input/provinces_code.csv"
PC_df=pd.read_csv(path_PC, encoding="ISO-8859-1")
PC_df["Province Code"]=PC_df["Province Code"].astype(str)

# Merge province code to 2011 dataframe

merged_2011_df=cwb2011_df.merge(PC_df, how='left', on="Province Code")
merged_2011_df.head()
```

```python
#Show just the header+5 lines
cwb2011_loc_df=merged_2011_df.loc[:, ["Province Code","Province","CSD Code / Code de la SDR 2011","CSD Name / Nom de la SDR 2011", "Census Population / Population de recensement 2011","CWB / IBC 2011", "Community Type / Type de collectivité 2011" ]]
cwb2011_final_df=cwb2011_loc_df.rename(columns={"CSD Code / Code de la SDR 2011":"CSD Code"})
cwb2011_final_df.head()
```

```python
# Make a reference to the books.csv file path
path_2006="../Input/cwb_2006.csv"

# Import the books.csv file as a DataFrame
cwb2006_df=pd.read_csv(path_2006, encoding="ISO-8859-1")

#Show just the header+5 lines
cwb2006_loc_df=cwb2006_df.loc[:, ["CSD Code / Code de la SDR 2006","CSD Name / Nom de la SDR 2006", "2A Population / Population 2A 2006","CWB / IBC 2006", "Community Type / Type de collectivité 2006" ]]
cwb2006_final_df=cwb2006_loc_df.rename(columns={"CSD Code / Code de la SDR 2006":"CSD Code"})
cwb2006_final_df["CSD Code"]=cwb2006_final_df["CSD Code"].astype(str)
cwb2006_final_df.head()
```

```python
# Make a reference to the books.csv file path
path_2001="../Input/cwb_2001.csv"

# Import the books.csv file as a DataFrame
cwb2001_df=pd.read_csv(path_2001, encoding="ISO-8859-1")

#Show just the header+5 lines
cwb2001_loc_df=cwb2001_df.loc[:, ["CSD Code / Code de la SDR 2001","CSD Name / Nom de la SDR 2001", "2A Population / Population 2A 2001","CWB / IBC 2001", "Community Type / Type de collectivité 2001" ]]
cwb2001_final_df=cwb2001_loc_df.rename(columns={"CSD Code / Code de la SDR 2001":"CSD Code"})
cwb2001_final_df["CSD Code"]=cwb2001_final_df["CSD Code"].astype(str)
cwb2001_final_df.head()
```

```python
# Make a reference to the books.csv file path
path_1996="../Input/cwb_1996.csv"

# Import the books.csv file as a DataFrame
cwb1996_df=pd.read_csv(path_1996, encoding="ISO-8859-1")

#Show just the header+5 lines
cwb1996_loc_df=cwb1996_df.loc[:, ["CSD Code / Code de la SDR 1996","CSD Name / Nom de la SDR 1996", "2A Population / Population 2A 1996","CWB / IBC 1996", "Community Type / Type de collectivité 1996" ]]
cwb1996_final_df=cwb1996_loc_df.rename(columns={"CSD Code / Code de la SDR 1996":"CSD Code"})
cwb1996_final_df["CSD Code"]=cwb1996_final_df["CSD Code"].astype(str)
cwb1996_final_df.head()
```

```python
merged_cwb1=cwb2011_final_df.merge(cwb2006_final_df, how='left', on="CSD Code")
merged_cwb2=merged_cwb1.merge(cwb2001_final_df, how='left', on="CSD Code")
merged_cwb3=merged_cwb2.merge(cwb1996_final_df, how='left', on="CSD Code")
merged_cwb_final=merged_cwb3.loc[:, ["Province Code","Province",'CSD Code', 'CSD Name / Nom de la SDR 2011',
       'Census Population / Population de recensement 2011', 'CWB / IBC 2011',
       'Community Type / Type de collectivité 2011', 
       '2A Population / Population 2A 2006', 'CWB / IBC 2006', 
       '2A Population / Population 2A 2001', 'CWB / IBC 2001', 
       '2A Population / Population 2A 1996', 'CWB / IBC 1996']]
merged_cwb_final
merged_cwb_final.head()
```

```python
merged_cwb_final=merged_cwb_final.replace('*',np.nan)
```

```python
merged_cwb_final=merged_cwb_final.replace('',np.nan)
```

```python
merged_cwb_final=merged_cwb_final.replace(' ',np.nan)
```

```python
droped_merged_cwb_final=merged_cwb_final.dropna(subset=['CWB / IBC 2011', 'CWB / IBC 2006',"CWB / IBC 2001", "CWB / IBC 1996"])
droped_merged_cwb_final.head()
```

```python
#Changing CWB to int
droped_merged_cwb_final["CWB / IBC 2011"]=droped_merged_cwb_final["CWB / IBC 2011"].astype(int)
droped_merged_cwb_final["CWB / IBC 2006"]=droped_merged_cwb_final["CWB / IBC 2006"].astype(int)
droped_merged_cwb_final["CWB / IBC 2001"]=droped_merged_cwb_final["CWB / IBC 2001"].astype(int)
droped_merged_cwb_final["CWB / IBC 1996"]=droped_merged_cwb_final["CWB / IBC 1996"].astype(int)
droped_merged_cwb_final["2A Population / Population 2A 1996"]=droped_merged_cwb_final["2A Population / Population 2A 1996"].astype('int64')
```

```python
#CALCULATING THE CWB INCREASE/DECREASE
droped_merged_cwb_final["CWB 2006/2011"]=(droped_merged_cwb_final["CWB / IBC 2011"]-droped_merged_cwb_final["CWB / IBC 2006"])/droped_merged_cwb_final["CWB / IBC 2011"]
droped_merged_cwb_final["CWB 2001/2006"]=(droped_merged_cwb_final["CWB / IBC 2006"]-droped_merged_cwb_final["CWB / IBC 2001"])/droped_merged_cwb_final["CWB / IBC 2006"]
droped_merged_cwb_final["CWB 1996/2001"]=(droped_merged_cwb_final["CWB / IBC 2001"]-droped_merged_cwb_final["CWB / IBC 1996"])/droped_merged_cwb_final["CWB / IBC 2001"]
droped_merged_cwb_final.head()
```

```python
#CALCULATING THE POPULATION INCREASE/DECREASE
droped_merged_cwb_final["POP 2006/2011"]=(droped_merged_cwb_final["Census Population / Population de recensement 2011"]-droped_merged_cwb_final["2A Population / Population 2A 2006"])/droped_merged_cwb_final["Census Population / Population de recensement 2011"]
droped_merged_cwb_final["POP 2001/2006"]=(droped_merged_cwb_final["2A Population / Population 2A 2006"]-droped_merged_cwb_final["2A Population / Population 2A 2001"])/droped_merged_cwb_final["2A Population / Population 2A 2006"]
droped_merged_cwb_final["POP 1996/2001"]=(droped_merged_cwb_final["2A Population / Population 2A 2001"]-droped_merged_cwb_final["2A Population / Population 2A 1996"])/droped_merged_cwb_final["2A Population / Population 2A 2001"]
droped_merged_cwb_final.head()
```

```python
growth_df=droped_merged_cwb_final.loc[droped_merged_cwb_final["Community Type / Type de collectivité 2011"].isin(['First Nations Communities','Inuit Communities'])] 

growth_df.sort_values("CWB 2001/2006", ascending=False)
```

```python
grouped_droped_merged_cwb_final=droped_merged_cwb_final.groupby(["Province Code", "Community Type / Type de collectivité 2011"]) 
droped_merged_cwb_final_sum=pd.DataFrame(grouped_droped_merged_cwb_final[
       '2A Population / Population 2A 1996', "index 1996",
       '2A Population / Population 2A 2001', "index 2001", 
       '2A Population / Population 2A 2006',"index 2006",
       'Census Population / Population de recensement 2011', "index 2011"
        ].sum())
droped_merged_cwb_final_sum
```

```python
droped_merged_cwb_final_sum["Province CWB 1996"]=droped_merged_cwb_final_sum["index 1996"]/droped_merged_cwb_final_sum["2A Population / Population 2A 1996"]
droped_merged_cwb_final_sum["Province CWB 2001"]=droped_merged_cwb_final_sum["index 2001"]/droped_merged_cwb_final_sum["2A Population / Population 2A 2001"]
droped_merged_cwb_final_sum["Province CWB 2006"]=droped_merged_cwb_final_sum["index 2006"]/droped_merged_cwb_final_sum["2A Population / Population 2A 2006"]
droped_merged_cwb_final_sum["Province CWB 2011"]=droped_merged_cwb_final_sum["index 2011"]/droped_merged_cwb_final_sum["Census Population / Population de recensement 2011"]
droped_merged_cwb_final_sum
```

```python
#Building CWB per province

CWB_province=droped_merged_cwb_final_sum.loc[:,["Province CWB 1996","Province CWB 2001","Province CWB 2006","Province CWB 2011"]]
CWB_province_rename=CWB_province.rename(columns={"Province CWB 1996":1996, 
                                              "Province CWB 2001":2001, 
                                              "Province CWB 2006":2006,
                                              "Province CWB 2011": 2011})
CWB_province_rename.to_csv("../Output/CWB_provinces_1996_2011.csv", 
                  encoding="utf-8", index=True, header=True)
```
