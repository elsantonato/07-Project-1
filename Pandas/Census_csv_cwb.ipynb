```python
#CALCULATING THE INDEX
droped_merged_cwb_final["index 2011"]=droped_merged_cwb_final["Census Population / Population de recensement 2011"]*droped_merged_cwb_final["CWB / IBC 2011"]
droped_merged_cwb_final["index 2006"]=droped_merged_cwb_final["2A Population / Population 2A 2006"]*droped_merged_cwb_final["CWB / IBC 2006"]
droped_merged_cwb_final["index 2001"]=droped_merged_cwb_final["2A Population / Population 2A 2001"]*droped_merged_cwb_final["CWB / IBC 2001"]
droped_merged_cwb_final["index 1996"]=droped_merged_cwb_final["2A Population / Population 2A 1996"]*droped_merged_cwb_final["CWB / IBC 1996"]
droped_merged_cwb_final["index 1981"]=droped_merged_cwb_final["2A Population / Population 2A 1981"]*droped_merged_cwb_final["CWB / IBC 1981"]

droped_merged_cwb_final.head()
```
