```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reference the file where the CSV is located
income_province_path = "../API/IncomePerProvince.csv"
income_reserve_path = "../API/IncomePerReserve.csv"

# Import the data into a Pandas DataFrame
income_province = pd.read_csv(income_province_path)
income_province.head()

income_reserve = pd.read_csv(income_reserve_path)
income_reserve.head()
```

```python
merged_df = pd.merge(income_province, income_reserve, on='Provinces', how='outer')
merged_df
```

```python
#drop canada
new_merged_df = merged_df.drop(merged_df.index[[0]])
#new_merged_df
final_merged_df = new_merged_df.sort_values(['Provinces'], ascending=[True])
final_merged_df
```

```python
merged_rename = final_merged_df.rename(columns={'Average Total Income_x': 'Average Total Income(Province)',
                                         'Average Total Income_y': 'Average Total Income(Reserve)'})
merged_rename
```

```python
# set width of bar

barWidth = 0.30

# set height of bar
province = [12778., 45616., 43767., 39141., 45210., 64586., 41479., 50689.,
47915., 38899., 42546., 49409., 53809.]

reserve = [24708.18229449, 33181.6761305 , 19281.03368242, 23526.05366357,
       36426.59782609, 31941. , 20698.14906832, 0,
       24568.00844695, 32564. , 26793.44109672, 20348.63305841,0]

 # Set position of bar on X axis
r1 = np.arange(len(province))
r2 = [x + barWidth for x in r1]


# Make the plot
plt.figure(figsize=(20, 10))  # width:20, height:3
ax1=plt.bar(r1, province, color='black', width=barWidth, edgecolor='white', label='Non-aboriginal', align='center')
ax2=plt.bar(r2, reserve, color='yellow', width=barWidth, edgecolor='white', label='Aboriginal',align='center')

# Add xticks on the middle of the group bars

plt.tick_params(labelsize=20)
plt.title('Average Income Per Province',fontsize=45,fontweight='bold')
plt.ylabel('Average Income', fontweight='bold',fontsize=18)
plt.xlabel('Province', fontweight='bold',fontsize=18)
plt.xticks([r + barWidth for r in range(len(province))], ['AB', 'BC', 'MB', 'NB',
                             'NL', 'NT',
                            'NS', 'NU', 'ON', 'PE',
                     'QC', 'SK', 'YT'], rotation = 0,fontweight='bold', fontsize=18)

plt.legend(loc='best', prop={'size': 25})
plt.savefig('../Output/Income.jpeg', bbox_inches='tight')
plt.show()
```
