```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
import csv

pr_csv = "../API/Population Growth Per Province.csv"
pr = pd.read_csv(pr_csv)
pr = pr.drop(0)
pr = pr.set_index('Provinces')
pr = pr [['Population, 2016','Population, 2011', 'Growth Rate' ]]
pr = pr.rename(columns = {'Population, 2016':'Population per Province, 2016','Population, 2011':'Population Per Province, 2011','Growth Rate':'Province Growth Rate'})
re_csv = "../API/Population Growth Per Reserve.csv"
re = pd.read_csv(re_csv)
re = re.set_index('Provinces')
re = re[['Population, 2016','Population, 2011', 'Growth Rate' ]]
re = re.rename(columns = {'Population, 2016':'Population per Reserve, 2016','Population, 2011':'Population Per Reserve, 2011','Growth Rate':'Reserve Growth Rate'})
growth = pd.merge(pr,re, on='Provinces')
#growth['Province Growth Rate'] = pd.Series(["{0:.2f}%".format(val) for val in growth['Province Growth Rate']], index = growth.index)
#growth['Reserve Growth Rate'] = pd.Series(["{0:.2f}%".format(val) for val in growth['Reserve Growth Rate']], index = growth.index)
growth
```

```python
growth = growth.sort_values(by=['Provinces'])
g = growth['Reserve Growth Rate'].plot(kind='bar')
#[plt.text(v, i, '{:.2f}%'.format(v),va='center') for i, v in enumerate(growth['Reserve Growth Rate'])]
plt.title('Reserve Population Growth Rate')
plt.xticks(rotation=70)
plt.xlabel('Provinces')
plt.ylabel('Growth Rate')   
plt.tight_layout()
plt.savefig("../Output/Reserve Population Growth Rate.png")
plt.show
```

```python
g = growth['Province Growth Rate'].plot(kind='bar')
#[plt.text(v, i, '{:.2f}%'.format(v),va='center') for i, v in enumerate(growth['Reserve Growth Rate'])]
plt.title('Province Population Growth Rate')
plt.xticks(rotation=70)
plt.xlabel('Provinces')
plt.ylabel('Growth Rate')   
plt.tight_layout()
plt.savefig("../Output/Province Population Growth Rate.png")
plt.show
```
