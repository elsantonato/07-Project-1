```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

male_csv = "../API/Male Population Age Breakdown Per Reserve.csv"
female_csv = "../API/Female Population Age Breakdown Per Reserve.csv"

male_pop = pd.read_csv(male_csv)
female_pop = pd.read_csv(female_csv)

# Data
provinces = list(male_pop)
provinces.remove('10')
male_pop = male_pop.replace(' years','',regex=True)
male_pop = male_pop.replace('to','-',regex=True)
male_pop = male_pop.replace('100 and over','>100',regex=True)
male_pop['10']
```

```python
for prov in provinces:
# Data
    ages = male_pop['10']
    male = male_pop[prov]
    female = female_pop[prov]
    y = np.arange(male.size)

# Sort by number of sales staff

    fig, axes = plt.subplots(ncols=2, sharey=True)
    axes[0].barh(y, male, align='center', color='blue', zorder=10)
    axes[0].set(title='Male')
    axes[1].barh(y, female, align='center', color='pink', zorder=10)
    axes[1].set(title='Female')

    axes[0].invert_xaxis()
#axes[0].set(yticks=y, yticklabels=ages)
#axes[0].yaxis.tick_right()
    axes[0].set(yticks=y, yticklabels=[])
    for yloc, age in zip(y, ages):
        axes[0].annotate(age, (0.48, yloc), xycoords=('figure fraction', 'data'), ha='center', va='center')

    for ax in axes.flat:
    #ax.margins(0.03)
        ax.grid(True)
    
    fig.tight_layout()
    fig.subplots_adjust(wspace=0.36)
    plt.suptitle(prov + " Province Age Pyramid",y=1.03)
    plt.savefig("../Output/" + prov + " Reserve Age Pyramid.png")
    plt.show
```
