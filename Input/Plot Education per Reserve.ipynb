```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

education_csv = "../API/EducationPerReserve.csv"

education = pd.read_csv(education_csv)

education = education.set_index('Provinces')
education = education.T

education['Canada']=education.sum(axis=1)
provinces = list(education)

education
```

```python
for province in provinces:
    colors = ["gold","lightblue","red"]
    explode = (0.1,0,0)
    plt.figure()
    plt.title(province + " Reserves Education Level")
    plt.pie(education[province], explode = explode, labels=education.index.values.tolist(), colors=colors,autopct="%1.1f%%", shadow=True, startangle=150)    
# Save Figure
    #plt.tight_layout()
    #plt.legend(loc = 'lower right')
    plt.savefig("../Output/" + province + " Reserves Education Level.png")
    
```
