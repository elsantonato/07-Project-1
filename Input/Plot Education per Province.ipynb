```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

education_csv = "../API/EducationPerProvince.csv"

education = pd.read_csv(education_csv)

education = education.set_index('Provinces')
education = education.T
provinces = list(education)

education
```

```python
for province in provinces:
    colors = ["gold","lightblue","red"]
    explode = (0,0,0.1)
    #plt.figure().set_size_inches(12.5, 8.5)
    plt.title(province + " Province Education Level")
    plt.pie(education[province], explode = explode, labels=education.index.values.tolist(), colors=colors,autopct="%1.1f%%", shadow=True, startangle=150)
# Save Figure
    #plt.tight_layout()
    plt.savefig("../Output/" + province + " Province Education Level.png")
    
```
