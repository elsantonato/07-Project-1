```python
import csv
import json
import matplotlib.pyplot as plt
import requests
import pandas as pd
from pprint import pprint
import numpy as np

response = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CR2016Geo.json?lang=E&geos=CSD&topic=1&&notes=0')

b = json.loads(response.text[2:])
pprint(b['DATA'])
```

```python
prov = []
reserve=[]
dguid=[]
for i in range((len(b['DATA']))):
    if b['DATA'][i][5] == 'Indian reserve':
        prov.append(b['DATA'][i][2])
        reserve.append(b['DATA'][i][4])
        dguid.append(b['DATA'][i][0])
```

```python
province_x_reserve = pd.DataFrame({'Province':prov,'Indian Reserve':reserve,'dguid':dguid})
province_x_reserve
```

```python
income = pd.DataFrame()
for id in province_x_reserve['dguid']:
    response2 = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CPR2016.json?lang=E&dguid='+id+'&topic=7&&notes=0')
    b2 = json.loads(response2.text[2:])
    #pprint(b2['DATA'])
    income = income.append(b2['DATA'])

income
```
