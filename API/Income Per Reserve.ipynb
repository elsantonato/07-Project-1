```python
import csv
import json
import matplotlib.pyplot as plt
import requests
import pandas as pd
from pprint import pprint
import numpy as np

response = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CR2016Geo.json?lang=E&geos=CSD&topic=1&&notes=0')

b = json.loads(response.text[2:])
#pprint(b['DATA'])
prov = []
dguid=[]
for i in range((len(b['DATA']))):
    if b["DATA"][i][5]=="Indian reserve":
        prov.append(b['DATA'][i][2])
        dguid.append(b['DATA'][i][0])

province_dguid = pd.DataFrame({'Province':prov,'dguid':dguid})
province_dguid

income = pd.DataFrame()
for id in province_dguid['dguid']:
    response2 = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CPR2016.json?lang=E&dguid='+id+'&topic=7&&notes=0')
    b2 = json.loads(response2.text[2:])
    #pprint(b2['DATA'])
    income = income.append(b2['DATA'])
    
income
```

```python
income_n = income.loc[income[10] == '  Number of total income recipients aged 15 years and over in private households - 25% sample data']
income_a = income.loc[income[10] == '    Average total income in 2015 among recipients ($)']
inc_n = pd.DataFrame({'Provinces':income_n[1],'dguid':income_n[2],'Number of Income Recipients':income_n[13]})#.set_index('Provinces')
inc_a = pd.DataFrame({'Provinces':income_a[1],'dguid':income_a[2],'Average Total Income':income_a[13]})#.set_index('Provinces')
res = pd.merge(inc_n,inc_a, on=['Provinces','dguid'])
res = res.replace('None',0)
res['Total Income'] = res['Number of Income Recipients']*res['Average Total Income']
inc = res.groupby('Provinces')['Number of Income Recipients','Total Income'].sum()
inc['Average Total Income'] = inc['Total Income']/inc['Number of Income Recipients']
inc
```

```python
inc.to_csv("IncomePerReserve.csv",encoding="utf-8", index=True, header=True)
income.to_csv("Income - Raw Data.csv",encoding="utf-8", index=False, header=True)
```
