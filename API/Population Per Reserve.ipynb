```python
import csv
import json
import matplotlib.pyplot as plt
import requests
import pandas as pd
from pprint import pprint
import numpy as np

response = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CR2016Geo.json?lang=E&geos=CSD&topic=1&&notes=0')

b = json.loads(response.text[2:])
#pprint(b['DATA'])
prov = []
dguid=[]
for i in range((len(b['DATA']))):
    if b["DATA"][i][5]=="Indian reserve":
        prov.append(b['DATA'][i][2])
        dguid.append(b['DATA'][i][0])

province_dguid = pd.DataFrame({'Province':prov,'dguid':dguid})
province_dguid

population = pd.DataFrame()
for id in province_dguid['dguid']:
    response2 = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CPR2016.json?lang=E&dguid='+id+'&topic=13&&notes=0')
    b2 = json.loads(response2.text[2:])
    #pprint(b2['DATA'])
    population = population.append(b2['DATA'])
    
population
```

```python
population_2016 = population.loc[population[8] == '1.1.1']
population_2011 = population.loc[population[8] == '1.1.2']
population_0014 = population.loc[population[8] == '1.2.2.1']
population_1465 = population.loc[population[8] == '1.2.2.2']
population_6500 = population.loc[population[8] == '1.2.2.3']

pop_2016 = pd.DataFrame({'Provinces':population_2016[1],'dguid':population_2016[2],'Population, 2016':population_2016[13]})
pop_2011 = pd.DataFrame({'Provinces':population_2011[1],'dguid':population_2011[2],'Population, 2011':population_2011[13]})
pop_0014 = pd.DataFrame({'Provinces':population_0014[1],'dguid':population_0014[2],'Population %, 0 to 14y':population_0014[13]})
pop_1465 = pd.DataFrame({'Provinces':population_1465[1],'dguid':population_1465[2],'Population %, 14 to 65y':population_1465[13]})
pop_6500 = pd.DataFrame({'Provinces':population_6500[1],'dguid':population_6500[2],'Population %, over 65y':population_6500[13]})

pop = pd.merge(pop_2016,pop_2011, on=['Provinces','dguid'])
pop = pd.merge(pop,pop_0014, on=['Provinces','dguid'])
pop = pd.merge(pop,pop_1465, on=['Provinces','dguid'])
pop = pd.merge(pop,pop_6500, on=['Provinces','dguid'])

pop = pop.replace('None',0)

pop['Pop0014'] = pop['Population, 2016']*pop['Population %, 0 to 14y']/100
pop['Pop1465'] = pop['Population, 2016']*pop['Population %, 14 to 65y']/100
pop['Pop6500'] = pop['Population, 2016']*pop['Population %, over 65y']/100
pop = pop.round({'Pop0014':0,'Pop1465':0,'Pop6500':0})
popu = pop.groupby('Provinces')['Population, 2016','Population, 2011'].sum()
popu['Growth Rate'] = ((popu['Population, 2016']/popu['Population, 2011'])-1)*100
popu
```

```python
popp = pop.loc[pop['Pop1465'] > 0]
popp = popp.groupby('Provinces')['Pop0014','Pop1465','Pop6500'].sum()
popp['Population %, 0 to 14y'] = 100*popp['Pop0014']/(popp['Pop0014']+popp['Pop1465']+popp['Pop6500'])
popp['Population %, 14 to 65y'] = 100*popp['Pop1465']/(popp['Pop0014']+popp['Pop1465']+popp['Pop6500'])
popp['Population %, over 65y'] = 100*popp['Pop6500']/(popp['Pop0014']+popp['Pop1465']+popp['Pop6500'])
popp = popp [['Population %, 0 to 14y','Population %, 14 to 65y','Population %, over 65y']]
popp = popp.round({'Population %, 0 to 14y':2,'Population %, 14 to 65y':2,'Population %, over 65y':2})
popp
```

```python
popf = pd.merge(popu,popp, on='Provinces')
popf
```

```python
population.to_csv("Population - Raw Data.csv",encoding="utf-8", index=True, header=True)
popf.to_csv("Population Growth Per Reserve.csv",encoding="utf-8", index=True, header=True)
```
