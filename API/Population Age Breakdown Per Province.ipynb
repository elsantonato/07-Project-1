```python
import csv
import json
import matplotlib.pyplot as plt
import requests
import pandas as pd
from pprint import pprint
import numpy as np

response = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CR2016Geo.json?lang=E&geos=PR&topic=1&&notes=0')

b = json.loads(response.text[2:])
#pprint(b['DATA'])
prov = []
dguid=[]
for i in range((len(b['DATA']))):
    prov.append(b['DATA'][i][2])
    dguid.append(b['DATA'][i][0])

province_dguid = pd.DataFrame({'Province':prov,'dguid':dguid})
province_dguid

population = pd.DataFrame()
for id in province_dguid['dguid']:
    response2 = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CPR2016.json?lang=E&dguid='+id+'&topic=13&&notes=0')
    b2 = json.loads(response2.text[2:])
    #pprint(b2['DATA'])
    population = population.append(b2['DATA'])
    
population
```

```python
pop = population.pivot_table(index=[1,4],columns=10,values=13,aggfunc=sum).reset_index()
pop = pop[[1,4,'    0 to 4 years','    5 to 9 years','    10 to 14 years','    15 to 19 years','    20 to 24 years','    25 to 29 years','    30 to 34 years','    35 to 39 years','    40 to 44 years','    45 to 49 years','    50 to 54 years','    55 to 59 years','    60 to 64 years','    65 to 69 years','    70 to 74 years','    75 to 79 years','    80 to 84 years','      85 to 89 years','      90 to 94 years','      95 to 99 years','      100 years and over']]
pop = pop.groupby(1).sum()
pop = pop.T
pop
```

```python
male = population.pivot_table(index=[1,4],columns=10,values=15,aggfunc=sum).reset_index()
male = male[[1,4,'    0 to 4 years','    5 to 9 years','    10 to 14 years','    15 to 19 years','    20 to 24 years','    25 to 29 years','    30 to 34 years','    35 to 39 years','    40 to 44 years','    45 to 49 years','    50 to 54 years','    55 to 59 years','    60 to 64 years','    65 to 69 years','    70 to 74 years','    75 to 79 years','    80 to 84 years','      85 to 89 years','      90 to 94 years','      95 to 99 years','      100 years and over']]
male = male.groupby(1).sum()
male = male.T
male
```

```python
fmale = population.pivot_table(index=[1,4],columns=10,values=17,aggfunc=sum).reset_index()
fmale = fmale[[1,4,'    0 to 4 years','    5 to 9 years','    10 to 14 years','    15 to 19 years','    20 to 24 years','    25 to 29 years','    30 to 34 years','    35 to 39 years','    40 to 44 years','    45 to 49 years','    50 to 54 years','    55 to 59 years','    60 to 64 years','    65 to 69 years','    70 to 74 years','    75 to 79 years','    80 to 84 years','      85 to 89 years','      90 to 94 years','      95 to 99 years','      100 years and over']]
fmale = fmale.groupby(1).sum()
fmale = fmale.T
fmale
```

```python
pop.to_csv("Population Age Breakdown Per Province.csv",encoding="utf-8", index=True, header=True)
male.to_csv("Male Population Age Breakdown Per Province.csv",encoding="utf-8", index=True, header=True)
fmale.to_csv("Female Population Age Breakdown Per Province.csv",encoding="utf-8", index=True, header=True)
```
