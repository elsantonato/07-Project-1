```python
import csv
import json
import matplotlib.pyplot as plt
import requests
import pandas as pd
from pprint import pprint
import numpy as np

response = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CR2016Geo.json?lang=E&geos=CSD&topic=2&&notes=0')

b = json.loads(response.text[2:])
#pprint(b['DATA'])

prov = []
dguid=[]
for i in range((len(b['DATA']))):
    if b["DATA"][i][5]=="Indian reserve":
            prov.append(b['DATA'][i][2])
            dguid.append(b['DATA'][i][0])

province_dguid = pd.DataFrame({'Province':prov,'dguid':dguid})
province_dguid

education = pd.DataFrame()
for id in province_dguid['dguid']:
    response2 = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CPR2016.json?lang=E&dguid='+id+'&topic=2&&notes=0')
    b2 = json.loads(response2.text[2:])
    #pprint(b2['DATA'])
    education = education.append(b2['DATA'])
    
education
```

```python
education_n = education.loc[education[8] == '10.1.1.1']
education_s = education.loc[education[8] == '10.1.1.2']
education_p = education.loc[education[8] == '10.1.1.3']
edu_n = pd.DataFrame({'Provinces':education_n[1],'No certificate, diploma or degree':education_n[13]})#.set_index('Provinces')
edu_s = pd.DataFrame({'Provinces':education_s[1],'Secondary (high) school diploma or equivalency certificate':education_s[13]})#.set_index('Provinces')
edu_p = pd.DataFrame({'Provinces':education_p[1],'Postsecondary certificate, diploma or degree':education_p[13]})#.set_index('Provinces')
edu_n = edu_n.replace('None',0)   
edu_s = edu_s.replace('None',0)   
edu_p = edu_p.replace('None',0)
edu_n = edu_n.groupby('Provinces').sum()
edu_s = edu_s.groupby('Provinces').sum()
edu_p = edu_p.groupby('Provinces').sum()
edu = edu_n.merge(edu_s, on='Provinces')
edu = edu.merge(edu_p, on='Provinces')
edu
```

```python
edu.to_csv("EducationPerReserve.csv",encoding="utf-8", index=True, header=True)
education.to_csv("Education - Raw Data.csv",encoding="utf-8", index=False, header=True)
```
