```python
import csv
import json
import matplotlib.pyplot as plt
import requests
import pandas as pd
from pprint import pprint
import numpy as np

response = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CR2016Geo.json?lang=E&geos=PR&topic=1&&notes=0')

b = json.loads(response.text[2:])
#pprint(b['DATA'])
prov = []
dguid=[]
for i in range((len(b['DATA']))):
    prov.append(b['DATA'][i][2])
    dguid.append(b['DATA'][i][0])

province_dguid = pd.DataFrame({'Province':prov,'dguid':dguid})
province_dguid

income = pd.DataFrame()
for id in province_dguid['dguid']:
    response2 = requests.get('https://www12.statcan.gc.ca/rest/census-recensement/CPR2016.json?lang=E&dguid='+id+'&topic=7&&notes=0')
    b2 = json.loads(response2.text[2:])
    #pprint(b2['DATA'])
    income = income.append(b2['DATA'])
    
income
```

```python
income_n = income[income[10].str.contains('  Number of total income recipients aged 15 years and over in private households - 25% sample data')]
income_a = income[income[10].str.contains('    Average total income in 2015 among recipients ($)')]
inc_n = pd.DataFrame({'Provinces':income_n[1],'Number of Income Recipients':income_n[13]}).set_index('Provinces')
inc_a = pd.DataFrame({'Provinces':income_a[1],'Average Total Income':income_a[13]}).set_index('Provinces')
inc = pd.merge(inc_n,inc_a, on='Provinces')
inc_a
```

```python
income.to_csv("income.csv",encoding="utf-8", index=False, header=True)
```
